```python
# Authentication process to use Tweepy API
consumer_key = "###################" #Your API/Consumer key 
consumer_secret = "###########################" #Your API/Consumer Secret Key
access_token = "##############################"    #Your Access token key
access_token_secret = "#################################" #Your Access token Secret key

#Pass in our twitter API authentication key
auth = tweepy.OAuthHandler(consumer_key,consumer_secret)
auth.set_access_token(access_token,access_token_secret)

#Instantiate the tweepy API
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser(),
                 wait_on_rate_limit=True,wait_on_rate_limit_notify=True)
```

```python
# Creating list of tweet ids
tweet_id = twitter_archive['tweet_id']
# Iterate over tweet_ids to obtain favorites and retweet count. Then append these has dicts into df_list.
df_list = []

# Any tweet_ids not found by API, are appended to e_list
e_list = []

for id in tweet_id:
    try:
        page = api.get_status(id, tweet_mode = 'extended')
        favorites = page['favorite_count']
        retweet_count = page['retweet_count']
        df_list.append({'tweet_id': int(id),
                        'favorites': int(favorites),
                        'retweet_count': int(retweet_count)
                       })
        print(id, favorites, retweet_count)
    
    except Exception as e:
        e_list.append(id)
```

```python
Twitter_APL_df = pd.DataFrame(df_list, columns = ['tweet_id', 'favorites', 'retweet_count'])
```

### Research Question 1: Whats are the top three breeds of dogs got the average highest retweet counts?

```python
#visualization 'Average retweet counts by the size of dog
df_size = df_clean[df_clean["Breed"].isnull()==False]
by_size = df_size.query('Breed=="pupper" or Breed=="doggo" or Breed=="puppo" or Breed=="floofer"')
retweet_count_by_sizes = pd.DataFrame(by_size.groupby('Breed').retweet_count.mean().sort_values())
retweet_count_by_sizes['Breed'] = retweet_count_by_sizes.index
retweet_count_by_sizes.plot(x="Breed",y='retweet_count',kind='barh',title="Average retweetcounts by Breed of dogs",color='r', alpha=0.6)
```

### Research Question 2: Which name of Dog are most tweeted about?

```python
#extract data that has retweet_count and save them into variable "retweet_and_favorite_count"
#get the index of ids with retweet_count
#create a column called 'predicted_breed' for putting the breed later on.

retweet_and_favorite_count = df_clean[df_clean["retweet_count"].isnull()==False]
id_with_retweet = list(retweet_and_favorite_count.tweet_id)
retweet_and_favorite_count['predicted_breed'] = None
```

```python
retweet_and_favorite_count.head()
```

```python
# find the ids that are also in the df_image.
has_data = []

for i in id_with_retweet:
    if len(list(img_prediction[img_prediction['tweet_id']==i].p1_dog.index))!=0:
        has_data.append(i)
    
len(has_data)
```

```python
# use for loop to determine their predicted breed by the highest probability of its predicted breed.

for i in has_data:
    index = list(retweet_and_favorite_count[retweet_and_favorite_count['tweet_id']==i].index)[0]
    if list(img_prediction[img_prediction['tweet_id']==i].p1_dog.values)[0] == True:
        retweet_and_favorite_count.predicted_breed[index] = list(img_prediction[img_prediction['tweet_id']==i].p1.values)[0]
    elif list(img_prediction[img_prediction['tweet_id']==i].p2_dog.values)[0] == True:
        retweet_and_favorite_count.predicted_breed[index] = list(img_prediction[img_prediction['tweet_id']==i].p2.values)[0]
    elif list(img_prediction[img_prediction['tweet_id']==i].p3_dog.values)[0] == True:
        retweet_and_favorite_count.predicted_breed[index] = list(img_prediction[img_prediction['tweet_id']==i].p3.values)[0]
```

```python
#test
retweet_and_favorite_count[retweet_and_favorite_count["predicted_breed"].isnull()==False].sample(5)
```

```python
# Plot a horizontal bar chart to show top 15 breeds
fig = plt.figure(figsize=(15,8))
retweet_and_favorite_count.groupby('predicted_breed').count()['rating_numerator'].sort_values(ascending=False).nlargest(5).plot(kind='barh')
plt.title("Top 5 Breeds of Dog Tweeted on WeRateDogs account",fontsize=20)
plt.ylabel("Dog Breed")
plt.xlabel("Number of Tweets") 
```

### Research Question 4: Whats are the top 5 underrated and overrated breed of Dogs ?

```python
#using quantile as division of retweet_count and ratings into different levels:
# min, mediun, moderatlyhigh, max
df_Quantile_breed = retweet_and_favorite_count[retweet_and_favorite_count["predicted_breed"].isnull()==False]
df_Quantile_breed.retweet_count.describe()
```

```python
df_Quantile_breed.rating_numerator.describe()
```

```python
#using the retweet and ratings .describe to divide into four levels(quantile) 

retweets = [11 ,516 ,1169 ,2690 ,69390] 
retweet_level = ['min', 'Medium','Moderately High' ,'max'  ]

df_Quantile_breed['retweet_levels'] = pd.cut(df_Quantile_breed['retweet_count'], retweets, labels=retweet_level)

ratings = [0 ,10 ,11 ,12 ,75] 
ratings_level = ['min', 'Medium','Moderately High' ,'max'  ]

df_Quantile_breed['rating_levels'] = pd.cut(df_Quantile_breed['rating_numerator'], ratings, labels=ratings_level)
```

```python
df_Quantile_breed.head(2)
```

```python
underrated_df = df_Quantile_breed.query('retweet_levels == "max" and rating_levels == "Medium" or retweet_levels == "Moderately High" and rating_levels == "min" or retweet_levels == "max" and rating_levels == "min" ')
overrated_df = df_Quantile_breed.query('retweet_levels == "min" and rating_levels == "Moderately High" or retweet_levels == "medium" and rating_levels == "max" or retweet_levels == "min" and rating_levels == "max"')
```

```python
# Visualize Top 5 most frequently undererated breeds by WeRateDogs

underrated = underrated_df.groupby('predicted_breed').tweet_id.count().sort_values(ascending=False).head(5)
underrated = pd.DataFrame(underrated)
underrated.rename(columns={'tweet_id':'underrated_counts'}, inplace=True)
underrated['predicted_breed'] = underrated.index
underrated.plot(x="predicted_breed",y='underrated_counts',kind='barh',title="Top 5 most frequently undererated breeds by WeRateDogs",color='r', alpha=0.6)
```

The top 5 most frequently underrated breeds by WeRateDogs are Golden Retriever, Labrador Retriver, pug, Pembroke, maiamute.
